# Filtering DataFrames with one Condition

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv',sep='\t')
```

```python
df.head()
```

```python
#setting the filter
filter_male = (df['Sex'] == 'male')
```

```python
#getting filtered rows using .LOC operator 
df.loc[filter_male]
```

```python
df.loc[df['Sex'] == 'female']
```

```python
#getting filtered rows, specifying the columns
df.loc[filter_male][['Name','Age']]
```

```python
df.dtypes
```

```python
##you can filter a DataFrame by data type
obj_types = df.dtypes == object
```

```python
df.loc[filter_male,obj_types]
```

```python
#use '~' character, if you want to exclude the specified data type 
df.loc[filter_male,~obj_types]
```
