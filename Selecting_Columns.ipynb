# Selecting Columns 

```python
import pandas as pd
```

```python
#get data from titanic.csv as set our DataFrame as df
df = pd.read_csv('titanic.csv','\t')
```

```python
df.head()
```

```python
#get all values from colunm 'Age'
df['Age']
```

```python
#the data type of our columns
type(df['Age'])
```

```python
#if we want a DataFrame instead of a pandas Series we have to do that
df[['Age']]
```

```python
type(df[['Age']])
```

```python
#selecting many columns
df[['Sex','Age','Fare']]
```

```python
#another way to select the column: you can pass the attribute 'dot' plus the name of the column
df.Age
```

```python
#in doubt if it is the same thing. You can check it out like this
df.Age.equals(df['Age'])
```
