# Slicing Rows and Columns with loc

```
import pandas as pd
```

```
arq = 'titanic.csv'
```

```
#read csv file (tab separated)
df = pd.read_csv(arq,sep='\t')
```

```
df.head()
```

```
#getting row by index
df.loc[2]
```

```
#getting the value of column 'Name' from row index 2
df.loc[2,'Name']
```

```
#getting rows from the begining to index 7
df.loc[:7]
```

```
#getting columns 'Name','Age' and 'Sex' from rows index 2 and 3
df.loc[[2,3],['Name','Age','Sex']]
```

```
#getting values from columns 'Name' and 'Age' from rows where column 'Survived' equals '1' 
df.loc[(df['Survived'] == 1),['Name','Age']]
```
