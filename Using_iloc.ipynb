# Extracting rows using Pandas .iloc

```python
import pandas as pd
```

```python
#getting data from csv file
df = pd.read_csv('titanic.csv',sep='\t')
```

```python
#checking the data
df.head()
```

```python
#get the firs row of the DataFrame
df.iloc[0]
```

```python
#notice that the type of the output is a Pandas Series
type(df.iloc[0])
```

```python
#get the last row of the DataFrame
df.iloc[-1]
```

### Slicing the data

```python
#getting the last 5 rows
df.iloc[-5:]
```

```python
#getting a slice of the data (note that the last index '126' is excluded from the search)
df.iloc[120:126]
```

```python
#getting specific rows
df.iloc[[5,67,150]]
```
